```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Load the dataset
data = pd.read_csv('../data/loan_data.csv')

# Basic information
print(data.info())
print(data.describe())

# Check for missing values
print(data.isnull().sum())

# Trim leading and trailing whitespace from column names
data.columns = data.columns.str.strip()

# Check for exact column names
print(data.columns)

# Ensure 'Loan_Status' column is correctly referenced
if 'loan_Status' in data.columns:
    # Correlation matrix (numeric columns only)
    numeric_columns = data.select_dtypes(include=['float64', 'int64']).columns
    plt.figure(figsize=(12, 8))
    sns.heatmap(data[numeric_columns].corr(), annot=True, cmap='coolwarm')
    plt.title('Correlation Matrix')
    plt.show()

    # Distribution of the target variable
    plt.figure(figsize=(8, 6))
    sns.countplot(x='loan_Status', data=data)
    plt.title('Loan Approval Status')
    plt.show()

    # Plotting some numeric features against the target variable
    plt.figure(figsize=(10, 6))
    sns.boxplot(x='loan_Status', y='Income_annum', data=data)  # Adjust y-axis variable as per your dataset
    plt.title('Income vs Loan Status')
    plt.show()

    plt.figure(figsize=(10, 6))
    sns.boxplot(x='loan_Status', y='Loan_amount', data=data)  # Adjust y-axis variable as per your dataset
    plt.title('Loan Amount vs Loan Status')
    plt.show()
else:
    print("Error: 'Loan_Status' column not found in dataset.")
```
